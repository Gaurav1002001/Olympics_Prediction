### This notebook file applies Ensemble Learning ( weighted average) to combine the three results from three models and derive one final result

```python
# Import required libraries and read test and train data set
import pandas as pd
import numpy as np
```

## Summer Dataset

#### Import the results of the three models

```python
linear_regression = pd.read_csv("data/Summer - Linear Regression Model.csv")
print(linear_regression.shape)
linear_regression.head()
```

```python
random_forest = pd.read_csv("data/Summer - Random Forest Model.csv")
print(random_forest.shape)
random_forest.head()
```

```python
CNN = pd.read_csv("data/Summer - Keras CNN Model.csv")
print(CNN.shape)
CNN.head()
```

```python
year = linear_regression["Year"]
true_value = list(linear_regression["True Value"])
NOC = linear_regression["Country"]

linear_predict = list(linear_regression["Predicted Value"])
random_predict = list(random_forest["Predicted Value"])
cnn_predict = list(CNN["Predicted Value"])
```

```python
Year = pd.DataFrame(year)
True_value = pd.DataFrame(true_value)
NOCs = pd.DataFrame(NOC)
```

```python
from sklearn.metrics import explained_variance_score
from sklearn.metrics import mean_squared_error
```

#### try different combination of weightage and choose the one with lowest MSE

```python
# try different combination of weightage and choose the one with lowest MSE
random_weights = [0.4, 0.5, 0.6, 0.7, 0.8]
performance_dict = {}

min_error = 100000
min_combination = "temp"
for random_weight in random_weights:
    linear_weights_heighest = 1.0 - random_weight
    linear_weights = [0.1*x for x in range((int)(linear_weights_heighest/0.1)) if 0.1*x < linear_weights_heighest]
    
    for linear_weight in linear_weights:
        CNN_weight = 1- random_weight - linear_weight
        NAME = "Linear {}- Random {}-CNN {}".format(linear_weight, random_weight, CNN_weight)
        
        print(NAME)
        
        loss = 0
        finalpred = []
        
        for i in range(196): # 196 countries to predict
            temp = int(round(linear_weight * linear_predict[i] + random_weight * random_predict[i] + CNN_weight * cnn_predict[i]))
            finalpred.append(temp)

        performance_dict[NAME] = mean_squared_error(true_value, finalpred) # total MSE of the prediction
        if (performance_dict[NAME] < min_error):
            print(performance_dict[NAME])
            min_error = performance_dict[NAME]
            min_combination = NAME

print()
print("Best:", min_combination)# ensemble learning using weighted average
```

```python
final_result_list = []
for i in range(196):
    print("{}/{}".format(i, 196))
    final_result_list.append(int(round(linear_predict[i] * 0.4 + random_predict[i] * 0.5 + cnn_predict[i] * 0.1)))
```

```python
final_result_summer = pd.DataFrame(final_result_list)
```

```python
final_summer = pd.concat([Year, NOCs, True_value, final_result_summer], axis = 1)
final_summer.columns = ["Year", "Country Code", "True Medals","Predicted Medals"]
```

```python
final_summer
```

```python
R2 = explained_variance_score(true_value, final_result_list)
MSE = mean_squared_error(true_value, final_result_list)
    
print("R2: ", R2, "MSE: ", MSE)
```

```python
final_summer.to_csv("Final Prediction Summer.csv", index = 0)
```
